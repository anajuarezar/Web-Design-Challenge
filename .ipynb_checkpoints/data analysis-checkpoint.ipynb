```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy import stats
```

```python
# We create the paths for our files
cities_path = "Resources/cities.csv"
```

```python
# We open it as csv file 
cities = pd.read_csv(cities_path)
cities.head()
```

```python
cities['Date'] = pd.to_datetime(cities['Date'], unit='s')
```

```python
cities['Date'] = cities['Date'].dt.strftime('%m/%d/%Y')
cities.head()
```

```python
plt.scatter(cities["Lat"], cities["Max Temp"], marker="o", facecolors="mediumorchid", edgecolors="thistle")
plt.xlim(-80, 100)
plt.ylim(-100, 150)
plt.grid()
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title('City Latitude vs Max Temperature (01/05/2017)')
plt.show()
```

```python
plt.scatter(cities["Lat"], cities["Humidity"], marker="o", facecolors="mediumorchid", edgecolors="thistle")
plt.xlim(-80, 100)
plt.ylim(-20, 120)
plt.grid()
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.title('City Latitude vs Humidity (01/05/2017)')
plt.show()
```

```python
plt.scatter(cities["Lat"], cities["Cloudiness"], marker="o", facecolors="mediumorchid", edgecolors="thistle")
plt.xlim(-80, 100)
plt.ylim(-20, 120)
plt.grid()
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')
plt.title('City Latitude vs Cloudiness (01/05/2017)')
plt.show()
```

```python
plt.scatter(cities["Lat"], cities["Wind Speed"], marker="o", facecolors="mediumorchid", edgecolors="thistle")
plt.xlim(-80, 100)
plt.ylim(-5, 40)
plt.grid()
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.title('City Latitude vs Wind Speed (01/05/2017)')
plt.show()
```
